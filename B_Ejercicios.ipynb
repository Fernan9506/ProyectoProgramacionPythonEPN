```python
# Para este proyecto se va utilizar pandas , numpy y matplotlib.pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 1 Lectura de Archivo de An√°lisis
path = "./forbes.csv"
datos = pd.read_csv(path)
datos.head()
```

```python
# 2 Ejercicio Listar Numero de Empresas de cada Sector
sectores=(datos.groupby("Sector").agg(Frecuencia=("Sector", "count")))
sectores.plot(kind='bar', title= "N√∫mero Empresas por Sector")
plt.show()
```

```python
# 3 Ejercicio Listar Numero de Empresas de cada Continente
sectores=(datos.groupby("Continent").agg(Frecuencia=("Continent", "count")))
sectores.plot(kind='bar', title= "N√∫mero Empresas por Continente")
plt.show()
```

```python
# 4 Listar la empresa top 1, adicional el promedio de todos los valores numericos, 
# suma del total de Ventas y Activos del ranking
print('****************** Empresa N√∫mero 1 *******************')
print(datos.loc[0])
print('*********** Medias de los Valores Num√©ricos ***********')
print(datos.mean())
print('****************** Ventas y Activos *******************')
print('Ventas Totales: ',sum(datos['Sales']), 'Billones de $')
print('Activos Totales: ',sum(datos['Assets']), 'Billones de $')
print('*************** Estad√≠stica de la Data ****************')
print(datos.describe())
```

```python
# 5 Listar todas las empresas que sus ventas sean mayores a 100 billones y menores a 150
datosVentas = pd.DataFrame(datos, columns=['Company','Industry','Continent','Country','Sales'])
datosVentas = datosVentas[(datosVentas.Sales>=100) & (datosVentas.Sales<=200)]
datosVentas
```

```python
# 6 Ejercicio Listar Numero de Empresas de cada Continente del Ejercicio Anterior
continentes=(datosVentas.groupby("Continent").agg(Frecuencia=("Continent", "count")))
print(continentes)
contin='Asia','Europa','Norte Am√©rica','Sur Am√©rica'
explode = (0.1, 0, 0.1, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(np.array(continentes.values).ravel(), explode=explode, labels=contin, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
# 7 Ejercicio Listar Numero de Empresas de cada Pais del Ejercicio Anterior
paises=(datosVentas.groupby("Country").agg(Frecuencia=("Country", "count")))
paises.plot(kind='bar')
plt.show()
```

```python
# 8 Graficar todas las ventas de estos paises
ventasGraf=(pd.DataFrame(datosVentas, columns=['Company','Sales']))
ventasGraf.plot(title="Gr√°fico de Ventas")
plt.show()
```
